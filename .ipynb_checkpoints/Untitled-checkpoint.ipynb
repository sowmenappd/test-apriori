```python
import pandas as pd
pd.set_option('display.max_colwidth',1)
```

```python
data = pd.read_csv("data.csv",error_bad_lines=False,header=None,warn_bad_lines=False)
data = data.iloc[:50]
data.apply(data.value_counts).sum(axis=1).sort_values(ascending=False)
data = data.drop(data.columns[0], axis=1)

headers = data.loc[0]
headers = [headers.iloc[x-1] for x in range(1, len(headers) + 1)]
headers
```

```python
for x in range(2, len(data)+1):
    data.loc[x-1] = data.loc[0] + '-' + data.loc[x-1]
```

```python
data
```

```python
records = []
for i in range((len(data))):
    records.append([str(data.values[i,j]) for j in range(0, data.shape[1])])
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(records).transform(records)
df = pd.DataFrame(te_data, columns = te.columns_)
# df = df.drop(("NaN"),axis=1)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori

df1 = apriori(df, min_support=.91, use_colnames = True, verbose=1)
df1.head()
```

```python
df1
```

```python
from mlxtend.frequent_patterns.association_rules import association_rules
```

```python
rules = association_rules(df1, metric = "confidence", min_threshold = 0.5)
rules
```
