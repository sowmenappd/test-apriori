```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

## Loading the main data CSV file

```python
df = pd.read_csv('data.csv',header=None)
```

## Get the JSON response file

```python
import json

file = open('./response_model.json', 'r')
options_model = json.load(file)['values']
file.close()
```

## Get all the question options as row elements

```python
options = []

for x in options_model:
    for y in x['options']:
        string = x['question'] + "-" + str(y)
        options.append(string)
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df = df.drop(index=0, axis=1)
```

```python
df.head()
```

```python
def get_column_header(answer, index):
    header = options_model[index]['question'] + "-" + answer
    return header
```

```python
data_rows = []

for index, row in df.iterrows():
    j = 1
    arr = []
    for col in row:
        arr.append(get_column_header(col, j))
        j += 1
    data_rows.append(arr)
```

```python
for data_row in data_rows:
    for opt in options:
        if opt not in data_row:
            data_row.append('0')
```

```python
matrix = []

for x in data_rows:
    entry = {}
    
    i = 1
    for _ in x:
        present_header = get_column_header(_, i)
        entry[present_header] = present_header
        i += 1
    
    matrix.append(entry)
```

```python
# matrix = []

# for i in range(0, len(data_rows)):
#     matrix.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
for x in matrix:
    present_keys = x.keys()
    
    for opt in options:
        if opt not in present_keys:
            x[opt] = 0
```

```python
processed = pd.DataFrame(matrix)
```

```python
processed.fillna(0, inplace=True)
processed.head(20)
```
